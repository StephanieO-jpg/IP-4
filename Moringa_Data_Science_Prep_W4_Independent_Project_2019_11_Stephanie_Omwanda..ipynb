# **Questions**

# **BLUECAR**

```
# Create new columns to display each car's usage
auto['bluecar_consumption'] = auto.bluecar.diff()
auto['utilib_consumption'] = auto.utilib.diff()
auto['utilib14_consumption'] = auto.utilib_14.diff()
```

```
# Identify the most popular hour of the day for picking up a shared electric car(Bluecar) in Paris in the month of April 2018
auto[(auto.city == 'Paris') & 
   (auto.bluecar_consumption < 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(1)
```

4 am is the most popular hour for picking a car

```
# Most popular hour for returning cars
auto[(auto.city == 'Paris') & 
   (auto.bluecar_consumption > 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(1)
```

9 pm (2100hrs) is the most popular hour for returning  the bluecar

```
# Most popular station overall
auto[(auto.station_kind == 'STATION') & 
   (auto.status == 'ok')].groupby('public_name')[['station_kind']].count().sort_values(by = 'station_kind', ascending  = False).head(1)
```

```
# Most popular station during the most popular picking hour
auto[(auto.station_kind == 'STATION') & (auto.hour == 4)  & 
   (auto.status == 'ok')].groupby('public_name').count()[['station_kind']].sort_values(by = 'station_kind', ascending = False).head(1)
```

```
# Most popular postal code for picking up Bluecars overall
auto[auto.bluecar_consumption < 0].groupby('postal_code').count()[['bluecar_consumption']].sort_values(by = 'bluecar_consumption', ascending = False).head(1)
```

```
# Does the most popular station belong to the most popular postal code
auto[auto.public_name == 'Paris/UzÃ¨s/1'][['public_name', 'postal_code']].head(1)
```

```
# Most popular postal code for picking up Bluecars at most most popular picking hours
auto[(auto.bluecar_consumption < 0) & 
   (auto.hour == 4)].groupby('postal_code').count()[['bluecar_consumption']].sort_values(by = 'bluecar_consumption', ascending = False).head(1)
```

```
# Does the most popular station belong to the most popular postal code for picking up bluecars at most popular hour
auto[(auto.public_name == 'Paris/Bosquet/41') & (auto.hour == 4)][['public_name', 'postal_code']].head(1)
```
